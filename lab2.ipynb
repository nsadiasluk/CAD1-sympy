## Derivative and integrate of functions

Computing the first and second derivative a functoion $f(x) = x^3 + 2x -1$

```python
x, y = symbols('x, y')
# Using function module from sympy 
f = Function('f')(x**3 + 2*x -1 )
f.diff()
```

```python
diff(f.diff())
```

```python
# using expressions 
f = x**3 + 2*x -1 
diff(f, x)
```

```python
diff(diff(f, x))
```

```python
diff(f, x, 2)
```

```python
# Defining the function using python 
def f(x):
    return x**3 + 2*x -1
f(x)
print('first derivative: ' , diff(f(x), x))
print('second derivative: ', diff(diff(f(x), x)))
```

Derivative of trigonometric functions $f(x) = \arctan(x)$

```python
f = atan(x)
```

```python
diff(f, x)
```

Integrating the results to get the back $f(x)$

```python
integrate(diff(f, x))
```

### Exercise:

```python
# Defining the function using python 
def f(x):
    return cos(x)
f(x)
print('first derivative: ' , diff(f(x), x))
print('integrate: ', integrate(diff(f(x), x)))

# Defining the function using python 
def g(x):
    return sin(x)
g(x)
print('first derivative: ' , diff(g(x), x))
print('integrate: ', integrate(diff(g(x), x)))
```

## Ploting 
### 2D plot

#### One plot:

```python
# One plot
u = symbols('u')
plot(sin(u), (u, 0, pi/2))
```

```python
plot(2*cos(x)-2+3*sin(2*x), (x, -pi, pi));
```

#### Plotting different expressions in one plot:

```python
# Different plots using one plot
plot(x**2, x**3, x**4-x+1,(x,-10,10))
```

#### Plotting different expresions using plot object:

```python
# plotting using the plot object 
p1 = plot(x**2, (x, -10, 10))
p2 = plot(x**3, (x, -10, 10))
```

#### p1 and p2 are plot objects, and have some properties like extend

```python
# p1 is a plot object 
print(p1)
```

```python
# which can be extended using extend 
p1.extend(p2)
```

```python
p1.show()
```

```python
plot(x**2, x**3, (x, -5, 5))
```

```python
c1 = 2*cos(x)-2+3*sin(2*x)
plot(cos(x), c1,(x, -pi, pi), legend=True)

```

#### colors in the plot
How to change the color in a plot, `line_color` properties of a plot

```python
plot(x**2, (x, -5, 5), line_color= 'green')
```

#### plotting multiple equations/function/expresions with different colors 
To plot different expressions with different colors, we can use the plot object to set the colors

```python
c1 = 2*cos(x)-2+3*sin(2*x)
p1 = plot(cos(x), c1,(x, -pi, pi), legend=True, show=False)
p1[0].line_color= 'r'
p1[1].line_color = 'b'
p1.show()
```

```python
p = plot((x**2,(x, -5, 5)), (x**3, (x, -5, 5)), show=False)
p[0].line_color = 'blue'
p[1].line_color = 'green'
p.show()
```

```python
f1 = x**2
f2 = -x**2
p = plot((f1, (x, -1, 0)), (f2, (x, 0, 1)), title="x**2 for [-1, 0), -x**2 for [0, 1]", show=False) 
p[0].line_color = 'r'
p[1].line_color = 'k'
p.show()
```

* Pay attention to the following to cells, we want to plot $\cos(nx)$ for $n \in [1, 3]$ in one plot

```python

l = [cos(n*x) for n in range(1, 4)]
p3 = plot(*l, (x, 0, 2*pi), show=False, title = '$\cos(nx)$ for n $\in$ [1, 3]', 
          legend=['cos(x)', 'cos(2x)', 'cos(3x)'])
p3[0].line_color = 'r'
p3[1].line_color = 'b'
p3[2].line_color = 'k'

p3.show()
```

## 3D plot
`Sympy.plotting` also provides the option to plot surfaces, lines in 3D

```python
x, y = symbols('x y')
plotting.plot3d((x*y)**2, (x, -5, 5), (y, -5, 5))
```

#### 3D lines

#### Surfaces

```python
plotting.plot3d_parametric_surface(cos(u+t), sin(u-t), u-t, (u, 0, 5), (t, 0, 5))
```
